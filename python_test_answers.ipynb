```python
# Q1 — Find Unique Elements
nums = [4,5,6,4,7,5,8]
result = [x for x in nums if nums.count(x) == 1]
print(result)
```

```python
# Q2 — Count Words in a Sentence
sentence = "python is great and python is easy"
words = sentence.split()

freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

print(freq)
```

```python
# Q3 — First Non-Repeating Character
s = "swiss"

for c in s:
    if s.count(c) == 1:
        print(c)
        break
```

```python
# Q4 — Rotate List by K Positions
lst = [1,2,3,4,5]
k = 2

k = k % len(lst)
rotated = lst[-k:] + lst[:-k]

print(rotated)
```

```python
# Q5 — Filter Even Numbers
nums = [11,22,33,44,55,66]

evens = [n for n in nums if n % 2 == 0]

print(evens)
```

```python
# Q6 — Count Occurrences of Elements
nums = [1,2,2,3,3,3,4]

freq = {}
for n in nums:
    freq[n] = freq.get(n, 0) + 1

print(freq)
```

```python
# Q7 — Remove Consecutive Duplicates
nums = [1,1,2,2,2,3,1,1,4]

result = [nums[0]]
for n in nums[1:]:
    if n != result[-1]:
        result.append(n)

print(result)
```

```python
# Q8 — Swap Case in a String
s = "PyThOn"
print(s.swapcase())
```

```python
# Q9 — Group Records by Key with Aggregation

data = [
 {"name":"Alice","subject":"Math","marks":90},
 {"name":"Bob","subject":"Math","marks":80},
 {"name":"Alice","subject":"Science","marks":85},
 {"name":"Bob","subject":"Science","marks":88},
 {"name":"Alice","subject":"English","marks":78}
]

result = {}

for d in data:
    name = d["name"]
    subject = d["subject"]
    marks = d["marks"]

    if name not in result:
        result[name] = {}

    result[name][subject] = marks

print(result)
```

```python
# Q10 — Clean and Standardize Product Data

data = [
 {"product":"  Apple iPhone 13 ", "price":70000},
 {"product":"samsung galaxy s21", "price":65000},
 {"product":"Apple iPhone 13", "price":70000},
 {"product":" Samsung Galaxy S21 ", "price":65000},
 {"product":"OnePlus 11", "price":60000}
]

seen = set()
result = []

for item in data:
    name = item["product"].strip().lower()

    if name not in seen:
        seen.add(name)
        result.append({
            "product": name,
            "price": item["price"]
        })

print(result)
```
