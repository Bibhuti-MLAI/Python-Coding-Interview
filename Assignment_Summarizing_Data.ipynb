# Problem 1

Write a function which calculates the mode of a list of numbers. If there are two modes in the dataset, it should return both.

```
from collections import Counter

def calculate_mode(numbers):
    """
    Calculates the mode(s) of a list of numbers.
    If there are two modes, it returns both.
    """
    if not numbers:
        return []

    counts = Counter(numbers)
    max_count = max(counts.values())

    modes = [key for key, value in counts.items() if value == max_count]

    # Sort for consistent output, especially for testing
    modes.sort()

    return modes
```

Let's test the function with a few examples:

```
# Example 1: Single mode
list1 = [1, 2, 2, 3, 4]
print(f"List: {list1}, Mode(s): {calculate_mode(list1)}")

# Example 2: Multiple modes (two modes)
list2 = [1, 2, 2, 3, 3, 4]
print(f"List: {list2}, Mode(s): {calculate_mode(list2)}")

# Example 3: All elements appear once (no distinct mode, all are modes)
list3 = [1, 2, 3, 4, 5]
print(f"List: {list3}, Mode(s): {calculate_mode(list3)}")

# Example 4: Empty list
list4 = []
print(f"List: {list4}, Mode(s): {calculate_mode(list4)}")

# Example 5: More than two modes (should return all)
list5 = [1, 1, 2, 2, 3, 3]
print(f"List: {list5}, Mode(s): {calculate_mode(list5)}")

# Example 6: Floating point numbers
list6 = [1.5, 2.0, 2.0, 3.5, 1.5]
print(f"List: {list6}, Mode(s): {calculate_mode(list6)}")
```
